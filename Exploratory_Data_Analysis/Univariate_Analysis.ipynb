```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

## Categorical Data

### a. Countplot

```python
sns.countplot(x='survived', data=df)
df['survived'].value_counts()
```

### b. PieChart

```python
df['sex'].value_counts().plot.pie(autopct='%1.1f%%')
```

## Numerical

### a. Histogram

```python
sns.histplot(df['age'])
```

```python
sns.histplot(df['age'],kde= True)
```

### b. Boxplot

```python
sns.boxplot(df['age'])
```

```python
sns.distplot(df['age'])
```
